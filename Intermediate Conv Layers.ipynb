```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import tensorflow as tf
from sklearn.datasets import load_sample_image
```

```python
china = load_sample_image('china.jpg') / 255
flower = load_sample_image('flower.jpg') / 255

images = np.array([china, flower])
```

```python
batch_size, height, width, channels = images.shape

filters = np.zeros(shape=(7,7,channels, 2), dtype=np.float32)
filters[:, 2:5, :, 0] = 1 # vertical line
filters[2:5, :, :, 1] = 1 # horizontal line
```

```python
outputs = tf.nn.conv2d(images, filters, strides=1, padding='SAME')
```

```python
print(outputs[0,:,:,1])
plt.imshow(outputs[1,:,:,1].eval(session=tf.compat.v1.Session()), cmap="gray") # plot 1st image's 2nd feature map
plt.show()
```

```python
plt.imshow(outputs[1,:,:,0].eval(session=tf.compat.v1.Session()), cmap="gray") # plot 1st image's 2nd feature map
plt.show()
```
